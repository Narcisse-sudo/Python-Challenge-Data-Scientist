```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
print ("Version de pandas:", pd.__version__)
print ("version python: ", np.__version__)
```

### Data

```python
df = sns.load_dataset("tips")
df.head()
```

### Analyse Univariée

 Variable Discrète

```python
df["sex"].value_counts(normalize=True, sort=True, ascending=False)
```

```python
# Avec pandas
df["sex"].value_counts().plot(kind="bar")
plt.show()
```

```python
# Avec seaborn

sns.countplot(df, x='sex')
plt.show()
```

### Variable Continue

Statistiques

```python
df['tip'].min()
df['tip'].max()
df['tip'].mean()

# avec descrebe pour une analyse complet
df["tip"].describe()
```

Histogramme

```python
# pandas
df['tip'].plot(kind='hist', bins =30)
plt.show()
```

```python
# seaborn
sns.displot(df, x = 'tip')
plt.show()
```

Boxplot

```python
fig, ax = plt.subplots(1,2, figsize =(10,4))

# seaborn
sns.boxplot(data=df, y='tip', ax=ax[0])

# matplolib
df['tip'].plot(kind='box', ax=ax[1])

plt.show()
```
