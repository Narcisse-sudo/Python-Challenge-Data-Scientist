```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.datasets import fetch_california_housing
import matplotlib.pyplot as plt
```

```python
X, y = fetch_california_housing(return_X_y=True)
```

```python
model = LinearRegression()
model.fit(X, y)
predict = model.predict(X)
predict
```

```python
y
```

```python
difference = predict - y
```

```python
plt.hist(difference, bins=25)
plt.show()
```

```python
mean_absolute_error(y_true=y, y_pred=predict)
```

```python
root_mean_squared_error(y_true=y, y_pred=predict)
```

```python
mean_squared_error(y_true=y, y_pred=predict)
```

# $R_{2}-score$

```python
np.random.seed(0)
x = np.random.rand(50,1) * 10
y = (3 * x - np.random.normal(0, 4, size=(50,1)) ).reshape(-1,1)

model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(x,y)
plt.plot(x, y_pred, c='green')
plt.show()
```

#### Calcul de SS_res

```python
for i in range(len(x)):
    plt.plot([x[i][0], x[i][0]], [y[i][0], y_pred[i][0]], color='red', linestyle="--", lw = 1)
plt.scatter(x, y, c='k', zorder = 1)
plt.plot(x, y_pred, c='green', label = 'Modele lineaire')
plt.show()
```

```python
SS_res = np.sum((y - y_pred)**2)
SS_res
```

#### calcul de SS_tot

```python
mean_y = np.mean(y) 
mean_y
```

```python
for i in range(len(x)):
    plt.plot([x[i][0], x[i][0]], [y[i][0], mean_y], color='blue', linestyle="--", lw = 1)
plt.scatter(x, y, c='k', zorder = 1)
plt.axhline(mean_y, c='green', label = 'Modele lineaire')
plt.show()
```

```python
SS_tot = np.sum((y-mean_y)**2)
SS_tot
```

#### R2-final

#

```python
1 - SS_res / SS_tot
```
